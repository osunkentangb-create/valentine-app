```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load dataset
data = pd.read_csv('/content/shuffled_data.csv')

print("Dataset shape:", data.shape)
print(data.head())

# Separate features and label
X = data.iloc[:, 0:3].values     # First 3 columns = features
y = data.iloc[:, 3].values       # Last column = label

print("X shape:", X.shape)
print("y shape:", y.shape)

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

print("Feature Min:", scaler.data_min_)
print("Feature Max:", scaler.data_max_)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Build ANN
model = keras.Sequential([
    layers.Dense(9, activation='sigmoid', input_shape=(3,)),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train
model.fit(X_train, y_train, epochs=300, verbose=1)

# Evaluate
loss, acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", acc)

# Extract first layer weights
W1, b1 = model.layers[0].get_weights()

print("\nW1 shape:", W1.shape)
print("b1 shape:", b1.shape)

print("\nW1:\n", W1)
print("\nb1:\n", b1)
```

```python
import numpy as np

# ==============================
# 1. Generate sine wave (Amplitude = 1)
# ==============================

N = 133                      # number of samples
fs = 8000                   # sampling frequency
f = 60                      # signal frequency

t = np.arange(N) / fs
signal = np.sin(2 * np.pi * f * t)   # sine wave (amplitude = 1)

# ==============================
# 2. Extract Features
# ==============================

dc = np.mean(signal)                # DC component (mean value)
mean_square = np.mean(signal**2)    # Mean Square
rms = np.sqrt(mean_square)          # RMS

print("------ RAW FEATURES ------")
print("DC :", dc)
print("Mean Square :", mean_square)
print("RMS :", rms)

# ==============================
# 3. Scale Features
# ==============================

sample = np.array([[rms, mean_square, dc]])

sample_scaled = scaler.transform(sample)

print("\n------ SCALED FEATURES ------")
print(sample_scaled)

# ==============================
# 4. Predict
# ==============================

prediction = model.predict(sample_scaled)

print("\n------ PREDICTION ------")
print("Output probability:", prediction[0][0])

if prediction[0][0] > 0.5:
    print("Predicted Class: 1 (Fault)")
else:
    print("Predicted Class: 0 (Normal)")
```

```python
import numpy as np

def test_signal(A, fs=8000, f=60, N=133):
    """
    Generate a sine signal with amplitude A, extract features, scale,
    and do a forward pass through the trained model.
    """

    # 1пёЏвѓЈ Generate signal
    t = np.arange(N) / fs
    signal = A * np.sin(2 * np.pi * f * t)

    # 2пёЏвѓЈ Extract features: RMS, DC, MS
    rms = np.sqrt(np.mean(signal**2))
    dc  = np.mean(signal)
    ms  = np.mean(signal**2)

    features = np.array([[rms, dc, ms]])

    print("\nAmplitude:", A)
    print("RAW FEATURES -> RMS: {:.5f}, DC: {:.5f}, MS: {:.5f}".format(rms, dc, ms))

    # 3пёЏвѓЈ Scale features
    features_scaled = scaler.transform(features)
    print("SCALED FEATURES ->", features_scaled)

    # 4пёЏвѓЈ Forward pass through model manually
    W1, b1 = model.layers[0].get_weights()
    W2, b2 = model.layers[1].get_weights()

    # Layer 1
    Z1 = np.dot(features_scaled, W1) + b1
    A1 = 1 / (1 + np.exp(-Z1))  # sigmoid activation
    print("Layer 1 -> Z1:", Z1, "A1:", A1)

    # Layer 2 (output)
    Z2 = np.dot(A1, W2) + b2
    A2 = 1 / (1 + np.exp(-Z2))  # sigmoid output
    print("Output -> Z2:", Z2, "A2 (probability):", A2)

    # 5пёЏвѓЈ Classification
    pred_class = 1 if A2 > 0.5 else 0
    print("Predicted class:", pred_class)
    return A2, pred_class

# Example tests
test_signal(1)
test_signal(5)
test_signal(10)
```
